# **Overview: Legal Scorecard Analysis**

Large enterprises (‚Äúclient‚Äù) often face a significant challenge in efficiently and objectively evaluating proposals from law firms responding to Requests for Proposals (RFPs). This process is often time-consuming, heavily reliant on manual review, and susceptible to subjective biases. The need to thoroughly assess each proposal against specific criteria, which can vary significantly across different RFPs, further complicates the selection process.

The core technical challenge involves developing a robust and adaptable AI system that can accurately score and provide reasoning for diverse legal firm proposals based on varying RFP requirements and client-defined scorecards. This involves:

1. **Processing Unstructured Text Data**: Legal proposals are typically submitted as lengthy, unstructured text documents (e.g., PDFs, Word documents). Extracting relevant information and key arguments from these documents in a structured and meaningful way for evaluation is a significant hurdle.

2. **Handling Variable Scorecard Categories and Guidelines**: Each RFP issued by the client can have a unique set of scorecard categories with specific weights and potentially accompanying scoring guidelines. The system needs to dynamically adapt to these varying criteria without requiring significant manual reconfiguration for each new RFP.

3. **Mapping Proposal Content to Scorecard Categories**: The AI system needs to intelligently map the information presented in the proposals to the specific categories defined in the scorecard. This requires understanding the nuances of legal language and the ability to identify relevant sections within the proposal that address each criterion.

4. **Generating Objective and Explainable Scores**: The system must not only produce a numeric score for each proposal against each category but also provide a clear, textual explanation of the reasoning behind the assigned score. This "explainability" is crucial for building trust in the AI-driven scoring process and allowing human reviewers to understand the LLM's assessment.

5. **Integrating with Existing SaaS Platform**: The AI-powered scoring engine needs to be seamlessly integrated into the existing legal RFP management SaaS platform without disrupting current workflows and ensuring data security and compatibility.

# **Understanding Inputs üî∞**

- **RFP document**: The RFP (Request for Proposal) document serves as the foundational communication issued by the enterprise client to potential legal service providers. It is a formal solicitation outlining the client's specific legal needs, project scope, required services, and the criteria by which submitted proposals will be evaluated. This document sets the stage for the entire proposal and evaluation process, ensuring that all bidding firms have a clear understanding of the client's expectations and the framework for their response.


- **Scorecard categories**: Scorecard categories represent the specific evaluation criteria defined by the enterprise client to assess the legal firm proposals. These categories are often weighted to reflect their relative importance in the client's decision-making process and can vary significantly from one RFP to another. They provide a standardized framework for evaluating each proposal consistently and objectively, ensuring that the selection process is based on the factors most critical to the client's needs and priorities. Examples of scorecard categories might include firm experience in similar cases, proposed team qualifications, understanding of the client's industry, proposed methodology, and pricing structure.


- **Proposals**: Proposals are the formal responses submitted by law firms in answer to the enterprise client's RFP document. These detailed submissions outline the law firm's qualifications, experience, proposed approach to the client's legal needs, team composition, fee structure, and any other information requested in the RFP. These proposals are the core documents that the client will evaluate against the predefined scorecard categories to determine which legal firm best aligns with their requirements and offers the most compelling solution. The quality and comprehensiveness of these proposals are crucial in the selection process.

# **Solution Architecture üí°**

Develop a **Legal Scorecard Analysis Agent**,  which is an autonomous multi-agent AI system powered by `Gemini` to score proposals based on RFP requirements and provide automated reasoning for each category score. Overall the AI system includes the following agents collaborating in a workflow:

1. **Elaborate Categories Agent**: The task of this agent is to elaborate the scoring category title and guideline, which will be used as essential information to score a proposal against scoring category

2. **Validate Categories Agent**: The task of this agent is to validate if each scoring category is clear or not and based on that decide whether to score a proposal against scoring category or not

3. **Proposal Scoring and Reasoning Agent**: The task of this agent is to finally provide a numeric score (on the scale of 0-10) and corresponding reasoning for a proposal against each scoring category



*For implementation, we will use **[Google Agent Development Kit (ADK)](https://google.github.io/adk-docs/)**, which is a flexible and modular framework for developing and deploying AI agents. ADK can be used with popular LLMs and open-source generative AI tools and is designed with a focus on tight integration with the Google ecosystem and Gemini models. ADK makes it easy to get started with simple agents powered by Gemini models and Google AI tools while providing the control and structure needed for more complex agent architectures and orchestration.*

## **5Ô∏è‚É£ Define Sub Agents**

## **Define Elaborate Categories Sub Agent ü§ñ**

````
############################################################
# After agent callback for Post processing of sub agent output
############################################################
def modify_elaborate_categories_agent_output(
    callback_context: CallbackContext,
) -> Optional[types.Content]:
    """
    Post processing the agent output to prepare input for next agent
    Returns new Content to *replace* the agent's original output
    """
    agent_name = callback_context.agent_name
    invocation_id = callback_context.invocation_id
    print(f"\n[Callback] Exiting agent: {agent_name} (Inv: {invocation_id})")

    current_state = callback_context.state.to_dict()
    user_input = json.loads(callback_context.user_content.parts[0].text)

    # user input fields
    rfp = user_input["rfp"]
    proposal = user_input["proposal"]
    scoring_categories = user_input["scoring_categories"]
    _scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in scoring_categories
    }

    # get fields from current state
    elaborated_scoring_categories = json.loads(
        current_state["elaborated_scoring_categories"]
    )
    for dict_ in elaborated_scoring_categories:
        category_id = dict_["category_id"]  # unique id of scoring category

        title = _scoring_categories_dict[category_id]["title"]
        dict_["title"] = title

    # prepare the input for the next agent (validate categories agent)
    next_agent_input = {
        "rfp": rfp,
        "proposal": proposal,
        "elaborated_scoring_categories": elaborated_scoring_categories,
    }
    return types.Content(
        parts=[types.Part(text=json.dumps(next_agent_input))],
        role="model",  # Assign model role to the overriding response
    )


#############################################################
# Define a elaborate_categories_agent as sub agent
#############################################################
elaborate_categories_agent = LlmAgent(
    name="ElaborateCategoriesAgent",
    model=MODEL_NAME,
    generate_content_config=GENERATE_CONTENT_CONFIG,
    instruction="""
    <TASK>
    You are an expert legal analyst with extensive experience in reviewing & interpreting RFP (Request for Proposals) document, submitted Proposals and elaborating scoring categories.
    Your primary goal is to elaborate the scoring categories title and guideline, which will be used as essential information
    to score proposals of different law Firms against an RFP.

    Given the following details of a RFP document, Proposal along with current scoring categories title and guideline (optional) which needs to be elaborated, provide a elaborated scoring category title
    and scoring category guideline for each category covering all the useful information so that title and guideline can be used clearly as the scoring requirement without any ambiguity.
    <END TASK>

    <RULES>
    1. Carefully review & interpret details in the RFP document and Proposal.
    2. Pay attention to category_id for each scoring category.
    3. Elaborate scoring category title and guideline for each category, which will be used as essential information to score proposals.
    <END RULES>

    <INPUT FORMAT>
    You will be provided RFP document, Proposal along with current scoring categories title and guideline (optional) which needs to be elaborated in a JSON format like
    {
       "rfp": "<rfp_document>",
       "proposal": "<proposal_as_reponse_to_rfp>",
       "scoring_categories": "<scoring_categories_with_title_guideline>"
    }
    <END INPUT FORMAT>

    <OUTPUT FORMAT>
    Respond ONLY with a JSON object matching this exact schema:
    Your output should be a list of dictionaries. Each value consists of the following information:
    {
      "category_id": "<the unique id of the scoring category>",
      "elaborated_scoring_category_title": "<an elaborated scoring category title in 1-2 sentences in english language>",
      "elaborated_scoring_category_guideline": "<an elaborated scoring category guideline in 2-3 sentences in english language>"
    }
    <END OUTPUT FORMAT>

    <TIPS>
    Only output the intended dictionary. Do not output any extra characters like "\" or "```" or unopened / unclosed bracket signs.
    Do not output any additional words or phrases other that the intended dictionary.
    <END TIPS>
    """,
    description="Elaborate the scoring category title and guideline",
    # Stores its output (the elaborated_scoring_categories) into the session state
    # under the key 'elaborated_scoring_categories'.
    output_key="elaborated_scoring_categories",
    after_agent_callback=modify_elaborate_categories_agent_output,  # Assign the callback here
)
````

## **Define Validate Categories Sub Agent ü§ñ**

````
############################################################
# After agent callback for Post processing of sub agent output
############################################################
def modify_validate_categories_agent_output(
    callback_context: CallbackContext,
) -> Optional[types.Content]:
    """
    Post processing the agent output to prepare input for next agent
    Returns new Content to *replace* the agent's original output
    """
    agent_name = callback_context.agent_name
    invocation_id = callback_context.invocation_id
    print(f"\n[Callback] Exiting agent: {agent_name} (Inv: {invocation_id})")

    current_state = callback_context.state.to_dict()
    user_input = json.loads(callback_context.user_content.parts[-1].text)

    # user input fields
    rfp = user_input["rfp"]
    proposal = user_input["proposal"]
    scoring_categories = user_input["scoring_categories"]
    _scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in scoring_categories
    }

    # get fields from current state
    elaborated_scoring_categories = json.loads(
        current_state["elaborated_scoring_categories"]
    )
    _elaborated_scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in elaborated_scoring_categories
    }

    validated_scoring_categories = json.loads(
        current_state["validated_scoring_categories"]
    )
    _validated_scoring_categories = []
    for dict_ in validated_scoring_categories:
        category_id = dict_["category_id"]  # unique id of scoring category

        is_scoring_category_clear = dict_["is_scoring_category_clear"]
        if is_scoring_category_clear.lower() == "yes":
            title = _scoring_categories_dict[category_id]["title"]
            elaborated_scoring_category_title = _elaborated_scoring_categories_dict[
                category_id
            ]["elaborated_scoring_category_title"]
            elaborated_scoring_category_guideline = _elaborated_scoring_categories_dict[
                category_id
            ]["elaborated_scoring_category_guideline"]

            _validated_scoring_categories.append(
                {
                    "category_id": category_id,
                    "title": title,
                    "elaborated_scoring_category_title": elaborated_scoring_category_title,
                    "elaborated_scoring_category_guideline": elaborated_scoring_category_guideline,
                    "is_scoring_category_clear": is_scoring_category_clear,
                }
            )
        else:
            _validated_scoring_categories.append(
                {
                    "category_id": category_id,
                    "title": title,
                    "elaborated_scoring_category_title": None,
                    "elaborated_scoring_category_guideline": None,
                    "is_scoring_category_clear": is_scoring_category_clear,
                }
            )

    # prepare the input for the next agent (proposal scoring and reasoning agent)
    next_agent_input = {
        "rfp": rfp,
        "proposal": proposal,
        "validated_scoring_categories": _validated_scoring_categories,
    }
    return types.Content(
        parts=[types.Part(text=json.dumps(next_agent_input))],
        role="model",  # Assign model role to the overriding response
    )


#############################################################
# Define a validate_categories_agent as sub agent
#############################################################
validate_categories_agent = LlmAgent(
    name="ValidateCategoriesAgent",
    model=MODEL_NAME,
    generate_content_config=GENERATE_CONTENT_CONFIG,
    instruction="""
    <TASK>
    You are an expert legal analyst with extensive experience in reviewing & interpreting RFP (Request for Proposals) document, submitted Proposals and validating if scoring category is clear or not based on the provided category title and guideline.
    Your primary goal is to validate each scoring category as `yes` or `no` representing if understanding of the
    scoring category is clear or not based on whether response in the proposal represents the RFP requirements for provided scoring category or not

    Given the following details of a RFP document, Proposal and elaborated scoring categories title and guideline which needs to be validated.
    Output `is_scoring_category_clear` as `yes` if scoring category is clear, otherwise `no`.
    <END TASK>

    <RULES>
    1. Carefully review and interpret details in the RFP document and Proposal.
    2. Pay attention to category_id for each scoring category.
    3. Validate each scoring category as `yes` or `no` representing if understanding of the scoring category is clear or not based on whether response in the proposal represents the RFP requirements for provided scoring category or not.
    <END RULES>

    <INPUT FORMAT>
    You will be provided RFP document, Proposal and along with elaborated scoring categories title and guideline in a JSON format like
    {
       "rfp": "<rfp_document>",
       "proposal": "<proposal_as_reponse_to_rfp>",
       "elaborated_scoring_categories": "<elaborated_scoring_categories_with_title_guideline>",
    }
    <END INPUT FORMAT>

    <OUTPUT FORMAT>
    Respond ONLY with a JSON object matching this exact schema:
    Your output should be a list of dictionaries. Each value consists of the following information:
    {
      "category_id": "<the unique id of the scoring category>",
      "is_scoring_category_clear": "<`yes` or `no` representing if understanding of scoring category is clear or not based on whether response in the proposal represents the RFP requirements for provided scoring category or not>"
    }
    <END OUTPUT FORMAT>

    <TIPS>
    Only output the intended dictionary. Do not output any extra characters like "\" or "```" or unopened / unclosed bracket signs.
    Do not output any additional words or phrases other that the intended dictionary.
    <END TIPS>
    """,
    description="Validate the scoring category title and guideline",
    # Stores its output (the validated_categories) into the session state
    # under the key 'validated_categories'.
    output_key="validated_scoring_categories",
    after_agent_callback=modify_validate_categories_agent_output,
)
````

## **Define Proposal Scoring and Reasoning Sub Agent ü§ñ**

````
#################################################################
# After agent callback for Post processing of sub agent output
#################################################################
def modify_proposal_scoring_reasoning_agent_output(
    callback_context: CallbackContext,
) -> Optional[types.Content]:
    """
    Post processing the final agent output
    """
    agent_name = callback_context.agent_name
    invocation_id = callback_context.invocation_id
    print(f"\n[Callback] Exiting agent: {agent_name} (Inv: {invocation_id})")

    current_state = callback_context.state.to_dict()
    user_input = json.loads(callback_context.user_content.parts[0].text)

    # user input fields
    # rfp = user_input["rfp"]
    # proposal = user_input["proposal"]
    scoring_categories = user_input["scoring_categories"]
    _scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in scoring_categories
    }

    # get fields from current state
    elaborated_scoring_categories = json.loads(
        current_state["elaborated_scoring_categories"]
    )
    _elaborated_scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in elaborated_scoring_categories
    }

    validated_scoring_categories = json.loads(
        current_state["validated_scoring_categories"]
    )
    _validated_scoring_categories_dict = {
        dict_["category_id"]: dict_ for dict_ in validated_scoring_categories
    }

    proposal_scoring_and_reasoning = json.loads(
        current_state["proposal_scoring_and_reasoning"]
    )
    for dict_ in proposal_scoring_and_reasoning:
        category_id = dict_["category_id"]  # unique id of scoring category

        title = _scoring_categories_dict[category_id]["title"]
        elaborated_scoring_category_title = _elaborated_scoring_categories_dict[
            category_id
        ]["elaborated_scoring_category_title"]
        elaborated_scoring_category_guideline = _elaborated_scoring_categories_dict[
            category_id
        ]["elaborated_scoring_category_guideline"]
        is_scoring_category_clear = _validated_scoring_categories_dict[category_id][
            "is_scoring_category_clear"
        ]

        dict_["title"] = title
        dict_["elaborated_scoring_category_title"] = elaborated_scoring_category_title
        dict_[
            "elaborated_scoring_category_guideline"
        ] = elaborated_scoring_category_guideline
        dict_["is_scoring_category_clear"] = is_scoring_category_clear

    return types.Content(
        parts=[types.Part(text=json.dumps(proposal_scoring_and_reasoning))],
        role="model",  # Assign model role to the overriding response
    )


#################################################################
# Define a proposal_scoring_and_reasoning_agent as sub agent
#################################################################
proposal_scoring_and_reasoning_agent = LlmAgent(
    name="ProposalScoringAndReasoningAgent",
    model=MODEL_NAME,
    generate_content_config=GENERATE_CONTENT_CONFIG,
    instruction="""
    <TASK>
    You are an expert legal analyst with extensive experience in reviewing & interpreting RFP (Request for Proposals) document, submitted Proposals and scoring/reasoning proposal against each scoring category.
    Your primary goal is to predict a numeric score (on the float point scale of 0.0 to 10.0) and corresponding reasoning against the scoring category title and guideline ONLY.

    Given the following details of a RFP document, Proposal and scoring categories title and guideline against which score/reasoning needs to be provided.
    Predict a `score` for submitted proposal by law firm on a float point scale of 0.0 to 10.0 based on the scoring category title and guideline
    along with a valid and proper explanation as a `reason` for chosing this score.
    <END TASK>

    <RULES>
    1. Carefully review and interpret details in the RFP document and Proposal.
    2. Pay attention to category_id for each scoring category.
    3. For the given RFP, score the submitted proposal by law firm on a float point scale of 0.0 to 10.0 based on the scoring category title and guideline
    along with a valid and proper explanation as a reason for chosing the score.

    Here are some details around score scale range:
    - Range (0.0-1.9): This score range is the worst, means submitted response in the proposal does not represent the RFP requirements for scoring category
    - Range (2.0-3.9): This score range is the bad, means submitted response in the proposal tiny bit represent the RFP requirements for scoring category
    - Range (4.0-6.9): This score range is the average, means submitted response in the proposal slighly represent the RFP requirements for scoring category
    - Range (7.0-9.0): This score range is the good, means submitted response in the proposal highly represent the RFP requirements for scoring category
    - Range (9.1-10.0): This score range is the excellent, means submitted response in the proposal completely represent the RFP requirements for scoring category
    3. For a specific scoring category, If the field `is_scoring_category_clear` is `no` then both score and reason MUST be null always
    <END RULES>

    <INPUT FORMAT>
    You will be provided RFP document, Proposal and along with validated scoring categories title and guideline in a JSON format like
    {
       "rfp": "<rfp_document>",
       "proposal": "<proposal_as_reponse_to_rfp>",
       "validated_scoring_categories": "<validated_scoring_categories_with_title_guideline>",
    }
    <END INPUT FORMAT>

    <OUTPUT FORMAT>
    Respond ONLY with a JSON object matching this exact schema:
    Your output should be a list of dictionaries. Each value consists of the following information:
    {
      "category_id": "<the unique id of the scoring category>",
      "ai_reason": "<a short valid and proper 2-3 sentences explanation as a reason in english language for chosing the score",
      "ai_score": "<a float point variable score on a 0.0 to 10.0 scale based on the above reason>"
    }
    <END OUTPUT FORMAT>

    <TIPS>
    Only output the intended dictionary. Do not output any extra characters like "\" or "```" or unopened / unclosed bracket signs.
    Do not output any additional words or phrases other that the intended dictionary.
    If the output is null do not convert it into any other datatype. When outputting `null`, do not add quotation marks and output only the word null.
    <END TIPS>
    """,
    description="Proposal scoring and reasoning against the scoring category",
    # Stores its output (the proposal_scoring_and_reasoning) into the session state
    # under the key 'proposal_scoring_and_reasoning'.
    output_key="proposal_scoring_and_reasoning",
    after_agent_callback=modify_proposal_scoring_reasoning_agent_output,
)
````

## **6Ô∏è‚É£ Finally Define Legal Scorecard Analysis Agent ü§ñ**

#### This agent orchestrates the pipeline by running the sub agents in sequential order.

````
legal_scorecard_analysis_agent = SequentialAgent(
    name="LegalScorecardAnalysisAgent",
    sub_agents=[
        elaborate_categories_agent,
        validate_categories_agent,
        proposal_scoring_and_reasoning_agent,
    ],
)
````

````
# @title Define Agent Interaction Logic
async def call_agent_and_print(
    runner_instance: Runner, agent_instance: LlmAgent, session_id: str, query_json: str
):
    """Sends a query to the specified agent/runner and prints results."""
    print(f"\n>>> Calling Agent: {agent_instance.name}")

    user_content = types.Content(role="user", parts=[types.Part(text=query_json)])
    final_response_content = "No final response received."
    async for event in runner_instance.run_async(
        user_id=USER_ID, session_id=session_id, new_message=user_content
    ):
        # print(f"Event: {event.type}, Author: {event.author}") # Uncomment for detailed logging
        if event.is_final_response() and event.content and event.content.parts:
            # For output_schema, the content is the JSON string itself
            final_response_content = event.content.parts[0].text
    try:
        # Attempt to parse and pretty print if it's JSON
        final_response_content = json.loads(final_response_content)
        final_response_content = json.dumps(final_response_content, indent=2)
    except (json.JSONDecodeError, TypeError):
        pass

    print(
        f"\n<<< Agent '{agent_instance.name}' \n\nFinal Response:\n {final_response_content}"
    )
````
