## Overview

### Imagen 4

Imagen 4 on Vertex AI brings Google's state-of-the-art generative AI capabilities to application developers. It's capable of creating images with astonishing detail and rendered text across a wide variety of styles. With Imagen 4, you can also upscale a generated, edited, or existing image to increase the size of an image without losing quality.


In this tutorial, you will learn how to use the Google Gen AI SDK for Python to interact with Imagen 4's upscaling model to increase images from:

- Imagen outputs
- Local files
- Cloud Storage

### Set Google Cloud project information and create client

To get started using Vertex AI, you must have an existing Google Cloud project and [enable the Vertex AI API](https://console.cloud.google.com/flows/enableapi?apiid=aiplatform.googleapis.com).

Learn more about [setting up a project and a development environment](https://cloud.google.com/vertex-ai/docs/start/cloud-environment).

```
import os

PROJECT_ID = "[your-project-id]"  # @param {type: "string", placeholder: "[your-project-id]", isTemplate: true}
if not PROJECT_ID or PROJECT_ID == "[your-project-id]":
    PROJECT_ID = str(os.environ.get("GOOGLE_CLOUD_PROJECT"))

LOCATION = os.environ.get("GOOGLE_CLOUD_REGION", "us-central1")

client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
```

### Load the Imagen 4 models

```
generation_model = "imagen-4.0-generate-001"
upscale_model = "imagen-4.0-upscale-preview"
```

## Upscale images generated with Imagen

With Imagen 4, you can generate high quality images with the following parameters:
- `aspect ratio`: `1:1`, `9:16`, `16:9`, `3:4`, `4:3`
- `number_of_images`: any value between 1 and 4
- `image_size`: `1K` or `2K`

Once generated, you can then upscale it by a factor of `x2`, `x3`, or `x4`.

By default, a digital watermark, or [SynthID](https://deepmind.google/technologies/synthid/), is added to Imagen 4 images. You can [verify a watermarked image](https://console.cloud.google.com/vertex-ai/studio/media/) via Vertex AI Studio.

```
prompt = "A cartoon logo of a cat in a chef hat"

image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        aspect_ratio="1:1",
        number_of_images=1,
        image_size="2K",
    ),
)

upscale = client.models.upscale_image(
    model=upscale_model,
    image=image.generated_images[0].image,
    upscale_factor='x2',
)
upscale.generated_images[0].image.show()
```

## Upscale local images

You're also able to upscale local images with Imagen 4. In the next cell you'll download an image to use in the request. If you'd like to use a different image, modify the `image` variable below.

```
!wget https://storage.googleapis.com/cloud-samples-data/generative-ai/image/boats.jpeg

image = "boats.jpeg"
```

```
upscale = client.models.upscale_image(
    model=upscale_model,
    image=types.Image.from_file(location=image),
    upscale_factor='x3',
)
upscale.generated_images[0].image.show()
```

## Upscale images from Cloud Storage

Additionally, you can upscale images stored in Cloud Storage. If you'd like to change the initial image in this example modify the `image` variable below.

```
image = "gs://cloud-samples-data/generative-ai/image/daisy.jpg"

upscale = client.models.upscale_image(
    model=upscale_model,
    image=types.Image(gcs_uri=image),
    upscale_factor='x4',
)
upscale.generated_images[0].image.show()
```
